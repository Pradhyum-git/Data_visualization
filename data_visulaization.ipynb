## Data Visulaization :

```python
import pandas as pd
```

```python
df=pd.read_csv('/home/rgukt/Datasets/superstore.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Order_date']=pd.to_datetime(df['Order Date'],format='mixed')
```

```python
df['Ship_date']=pd.to_datetime(df['Ship Date'],format='mixed')
```

```python
df.rename(columns={'Row ID':'row_id','Order ID':'order_id','Ship Mode':'ship_mode','Customer ID':'customer_id','Customer Name':'customer_name','Postal Code':'postal_code','Product ID':'product_id','Product Name':'product_name','Sub-Category':'sub_category'},inplace=True)
```

```python
df.info()
```

```python
df['year']=df['Order_date'].dt.year
```

```python
df['month']=df['Order_date'].dt.month_name()
```

```python
df.head()
```

```python
df.drop(['Order Date','Ship Date'],axis=1,inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.pie(df['month'].value_counts(),autopct='%0.01f%%',labels=df['month'].value_counts().index)
plt.show()
```

```python
sns.countplot(data=df,x='month')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.pie(df['year'].value_counts(),autopct='%0.01f%%',labels=df['year'].value_counts().index)
plt.show()
```

```python
sns.countplot(data=df,x='year')
plt.xticks(rotation=90)
plt.show()
```
